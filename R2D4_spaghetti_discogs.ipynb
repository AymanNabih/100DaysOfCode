```python
import pandas as pd
MS = pd.read_csv("masters.csv")
MS.head()
```

```python
newdf = MS.pivot_table(index = 'year', columns= 'artist', 
                aggfunc = "count")
```

```python
df2 = pd.DataFrame(newdf)
df2.head()
```

```python
df2.columns = df2.columns.droplevel()
df2.head()
```

```python
df2.reset_index(inplace = True)
df2.head()
```

```python
df3 = pd.DataFrame(df2.year)
df3.head()
```

```python
list(df2.columns.values)
```

```python
df4 = df2.drop(columns = ['year'])
```

```python
df4.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sampled = df4.sample(n=10000, axis=1)
```

```python
sampled.head()
```

```python
# style
def showfig(sampled):
    plt.style.use('seaborn-darkgrid')
    import random
 
    # create a color palette
    palette = plt.get_cmap('Set1')
 
    # multiple line plot
    num= range(0,10)
    for c in sampled:
        column = pd.Series(c)
        column.dropna()
        plt.plot(df3['year'], sampled[column], marker='', color=palette(random.choice(num)), linewidth=1, alpha=0.9,
                     label=column)
 

    # Add titles
    plt.title("Album Releases by Artist", loc='left', fontsize=12, fontweight=0, color='orange')
    plt.xlabel("Year")
    plt.ylabel("Frequency")
    return plt
```

```python
fig1 = showfig(sampled)
fig1.savefig('albumfreqN2year.png', dpi=120)
```
