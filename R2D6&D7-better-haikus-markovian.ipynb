```python
from pyphen import Pyphen
def syllable_count(text, lang='en_US'):
    dic = Pyphen(lang=lang)
    count = 0
    for word in text.split(' '):
        word_hyphenated = dic.inserted(word)
        count += max(1, word_hyphenated.count("-") + 1)
    return count
```

```python
haikus = []
def makehaikus(n):

    for i in range(1, n):
        j = 0
        check = 0
        new = model_a.make_sentence()
        if new is not None:
            check = syllable_count(new, lang='en-US')
            print(check)
            if check == 17:
                j = j + 1
                print("yes!")
                haiku = str(new) + " "
                if haiku not in haikus:
                    haikus.append(haiku)   
            else:
                print("fail")
                print(new)
        if j == n:       
            return haikus
```

### here is a model that places weights the first model (Alice) twice as much as the second (Sherlock)

```python
model_combo = markovify.combine([ model_a, model_b ], [ 1.5, 0.75 ])
```

```python
mixedhaikus = []
def makehaikus(n):

    for i in range(1, n):
        j = 0
        check = 0
        new = model_combo.make_sentence()
        if new is not None:
            check = syllable_count(new, lang='en-US')
            print(check)
            if check == 17:
                j = j + 1
                print("yes!")
                haiku = str(new) + " "
                if haiku not in mixedhaikus:
                    mixedhaikus.append(haiku)   
            else:
                print("fail")
                print(new)
        if j == n:       
            return mixedhaikus
```
