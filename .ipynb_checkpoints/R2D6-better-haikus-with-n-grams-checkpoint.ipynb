### Find common bigrams between POS

```python
import collections

# Iterate through list and count bigram frequencies The return value is a dict
# mapping the length of the bi-gram to a collections. Counter object of n-gram tuple and 
# number of times that n-gram occurred. Returned dict includes n-grams of length min_length to max_length.

def count_bigrams(lines):
    lengths = range(2, 3)
    queue = collections.deque(maxlen=3)
    ngrams = {length: collections.Counter() for length in lengths}

    # Helper function to add bigrams at start of current queue to dict
    def add_queue():
        current = tuple(queue)
        for length in lengths:
            if len(current) >= length:
                ngrams[length][current[:length]] += 1

    # Loop through all lines and words and add bigrams to dict
    for line in lines:
            queue.append(line)
            if len(queue) >= 2:
                add_queue()

    # Make sure we get the bigrams at the tail end of the queue
    while len(queue) > 2:
        queue.popleft()
        add_queue()

    return (ngrams)
```

### Find common n-grams between POS

```python
import collections
def count_ngrams(lines, min_length=2, max_length=5):

    lengths = range(min_length, max_length + 1)
    ngrams = {length: collections.Counter() for length in lengths}
    queue = collections.deque(maxlen=max_length)

    # Helper function to add n-grams at start of current queue to dict
    def add_queue():
        current = tuple(queue)
        for length in lengths:
            if len(current) >= length:
                ngrams[length][current[:length]] += 1

    # Loop through all lines and words and add n-grams to dict
    for line in lines:
            queue.append(line)
            if len(queue) >= max_length:
                add_queue()

    # Make sure we get the n-grams at the tail end of the queue
    while len(queue) > min_length:
        queue.popleft()
        add_queue()

    return ngrams
```

```python
def haiku(n, df, df2):
    haikus = []
    haikuText = ""
    for i in range (1, n):
        C1 = random.choice(df.index)
        L1 = df[C1]
        C2 = random.choice(df2.index)
        L2 = df2[C2]
        C3 = random.choice(df.index)
        L3 = df[C3]
        haikuText = str(L1)+" "+ str(L2)+" " + str(L3)+" "
        haikus.append(haikuText)
    return haikus

        
        
    
```

```python
def writeHaikus(haikuList):
    with open("genHaikusBigrams.txt", "w") as fp:
        count = 0
        for item in haikuList:
            fp.write("%s\n" % item)
            fp.write("\n")
            count += 1
        print ("There are", count, "lines in your file.")
```
